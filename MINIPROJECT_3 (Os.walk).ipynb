```python
tree = os.walk('/mnt/HC_Volume_18315164/home-jupyter/jupyter-d-martynov/lesson_1_DM/WORK_FILES/data/')
data_files = []
target_data = 'data.csv'
for item in tree:
    files = item[-1]
    
    if len(files) > 0 and target_data in files:
        path = os.path.join(item[0], target_data)
        data_files.append(path)
```

```python
data_files
```

```python
data_files_df = []
for i in data_files:
    df = pd.read_csv(i, index_col=0)
    df['name'] = i.split('/')[-2]
    df['date'] = i.split('/')[-3]
    data_files_df.append(df)
    
    
```
